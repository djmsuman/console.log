## Importing the libraries

```python
import pymongo
from pymongo import MongoClient
import pandas as pd
import numpy as np
```

## Connecting to the local MongoDB database

- Instead of creating 100 random sample data, the data is imported to the database from a sample dataset via [GitHub](https://github.com/ivylabs/mongodb-databases-sample) provided by [IvyLabs](http://www.ivy-is.co.uk/).
- Using the `region` collection which has 110 data/objects/rows

```python
db = MongoClient("localhost", 27017).foodmart
```

## Fetching the data from database and create the dataframe

```python
df = pd.DataFrame(list(db.region.find()))
df
```

## Saving the data into a CSV file

```python
df.to_csv('regions.csv', index=False, encoding='utf-8')
```

## Reading the CSV file

```python
df = pd.read_csv("regions.csv")
```

## Adding a new column named "sale_index" at end of the dataframe with randomly generated values
